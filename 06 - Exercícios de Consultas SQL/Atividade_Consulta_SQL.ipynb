```python
!pip install mysql-connector-python

import mysql.connector
import pandas as pd

host = "SEU_HOST"
user = "SEU_USUARIO"
password = "SUA_SENHA"
database = "SEU_BANCO"

!curl ipecho.net/plain

def consulta(query, tabela):
    connection = mysql.connector.connect(
        host = host,
        user = user,
        password = password,
        database = database
    )
    cursor = connection.cursor()

    try:
        cursor.execute(query)
        result = cursor.fetchall()
        globals()[tabela] = pd.DataFrame(result, columns=cursor.column_names)
        display(globals()[tabela])
    finally:
        cursor.close()
        connection.close()
```

##### 50 - Liste os clientes, os carros que alugaram, as marcas desses carros, e o valor total gasto em aluguéis, filtrando por um intervalo de datas específico.
---

```python
consulta("""
SELECT
    cl.nome,
    GROUP_CONCAT(DISTINCT CONCAT
        (" ",ma.marca, " - ", ca.modelo)
        ORDER BY ca.modelo SEPARATOR " | ")
        AS carros_alugados,
        GROUP_CONCAT(DISTINCT al.data_aluguel
                 ORDER BY al.data_aluguel SEPARATOR " - ") AS datas_alugueis,
    SUM(ca.valor) soma_valores
FROM cliente cl
JOIN aluguel al
    ON al.codcliente = cl.codcliente
JOIN carro ca
    ON al.codcarro = ca.codcarro
JOIN marca ma
    ON ca.codmarca = ma.codmarca
WHERE al.data_aluguel
    BETWEEN "2023-04-01"
    AND "2023-04-30"
GROUP BY cl.nome;
""",
"""
soma_alugueis_entre_datas
""")
```
