```python
import pandas as pd
import math

df = pd.read_csv("https://storage.googleapis.com/atividade_aula_06/dados-brutos/dados_inconsistentes.csv")

df.info()

nomes = df["nome"].tolist()
idades = df["idade"].tolist()
salarios = df["salário"].tolist()
cargos = df["cargo"].tolist()
departamentos = df["departamento"].tolist()

nomes_tratados = []
idades_tratadas = []
salarios_tratados = []
cargos_tratados = []
departamentos_tratados = []

for i in range(len(nomes)):
    nome = nomes[i]
    idade = idades[i]
    salario = salarios[i]
    cargo = cargos[i]
    departamento = departamentos[i]

    if (nome not in [None, "erro", "Desconhecido"] and
        cargo not in [None, "erro", "Desconhecido"] and
        departamento not in [None, "erro", "Desconhecido"] and
        not (isinstance(idade, float) and math.isnan(idade)) and
        idade != "erro" and idade != "Desconhecido" and
        not (isinstance(salario, float) and math.isnan(salario)) and
        salario != "erro" and salario != "Desconhecido"):

        nomes_tratados.append(nome)
        idades_tratadas.append(int(idade) if isinstance(idade, float) else idade)
        salarios_tratados.append(salario)
        cargos_tratados.append(cargo)
        departamentos_tratados.append(departamento)

df_tratado = pd.DataFrame({
    "nome": nomes_tratados,
    "idade": idades_tratadas,
    "salário": salarios_tratados,
    "cargo": cargos_tratados,
    "departamento": departamentos_tratados
})

display(df_tratado)
```

```python
df_tratado.to_csv("dados_tratados.csv", index=False)
```
